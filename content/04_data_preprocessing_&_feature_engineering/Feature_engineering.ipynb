# 1. Missing Value Imputation

### 1.0. Identifying columns with null values: issnull()
The method "isnull()" is applied to a column in a dataframe and returns an array of bool, indicating whether each record is null or not

```python
# For example, consider the column "MiscFeature"
df["MiscFeature"].isnull()
```

Now we would like to identify those columns which contain null values.

We use the mean() method to calculate the mean of the 0's and 1's. If the mean is greater than 0, the column contains at least one null value 

```python
null_cols = [col for col in df.columns if df[col].isnull().mean() > 0] 
null_str_cols = [col for col in str_cols if df[col].isnull().mean() > 0] 
null_num_cols = [col for col in int_cols + flt_cols if df[col].isnull().mean() > 0]
null_bool_cols = [col for col in bool_cols if df[col].isnull().mean() > 0]  
```

```python
print(null_cols)
```

```python
print("Number of Columns with Null values:", len(null_cols))
print("Percentage of null values:")
for col in null_cols:
    print(col, round(df[col].isnull().mean(), 4), ", Data Type: ", df[col].dtype)
```

There are different methods for missing data imputation. 
To implement these methods, we create two subsets of the set of null columns (null_cols) with less than 1% null values (null5_cols) and more than 1% null values (null99_cols).

```python
# Create a sublist of null_cols with less than 0.28% null values
null1_cols = [col for col in null_cols if df[col].isnull().mean() <0.028] 
null1_str_cols = [col for col in null_str_cols if df[col].isnull().mean() <0.028]
null1_num_cols = [col for col in null_num_cols if df[col].isnull().mean() <0.028] 
null1_bool_cols = [col for col in null_bool_cols if df[col].isnull().mean() <0.028] 
#Columns with more than 0.28 percent null values.
null99_cols = list(set(null_cols) - set(null1_cols)) 
null99_str_cols = list(set(null_str_cols) - set(null1_str_cols)) 
null99_num_cols = list(set(null_num_cols) - set(null1_num_cols)) 
null99_bool_cols = list(set(null_bool_cols) - set(null1_bool_cols)) 
```

### 2.1. Brute Force Method

```python
col_corr = set()
corr_matrix = X_train.corr()

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > 0.65: 
            print(abs(corr_matrix.iloc[i, j]), corr_matrix.columns[i], corr_matrix.columns[j])
            colname = corr_matrix.columns[j]
            col_corr.add(colname)
```
